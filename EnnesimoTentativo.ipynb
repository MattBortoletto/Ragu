```python
import random
import numpy as np

def DNAgen(length):
    return ''.join(random.choice('CGTA') for _ in range(length))

print(DNAgen(5))
```

Idea:
- scelgo la lunghezza complessiva della sequenza di DNA
- genero le parole importanti nel giusto numero
- genero le restanti parole random in modo da arrivare al numero scelto inizialmente
- faccio uno shuffle tenendo fisso il numero di parole

```python
DNA_imp = [None]*51

for i in range(51):
    DNA_imp[i] = np.random.choice(["CCGCG", "CGCGC", "GCGCC", "CGGCG", "CGGCC"], 
                            p=[0.10554704, 0.11982444, 0.31040135, 0.10084078, 0.36338639])
```

```python
len_dna = 500000
DNAc = [None]*int((len_dna-len(DNA_imp))/5) # DNA casuale
print(len(DNAc))

for i in range(len(DNAc)):
    DNAc[i] = DNAgen(5)
```

```python
DNA_complete = DNA_imp + DNAc
```
